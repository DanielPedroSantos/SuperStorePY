```python
 # Importando dados 

import pandas as pd 
from sqlalchemy import text
from sqlalchemy import create_engine
from dotenv import load_dotenv
from utils.acessobanco import acessobanco
```

```python
# Cria a conexão com o banco usando a função acessobanco()

engine = acessobanco()
```

```python
# criando o dataframe puxando do banco da tablea orders na stg

df = pd.read_sql_table("orders",engine, schema="stg")
```

```python
# Alterando o tipo das colunas data_pedido e data_envio para tipo datetime

df[['data_pedido', 'data_envio']] = df[['data_pedido', 'data_envio']].apply(pd.to_datetime, errors='coerce')
```

```python
# Realizando uma conferencia dos tipos apos o ajuste

df.info()
```

```python
# Verificando apenas as colunas de datas com um top 5

df[['data_pedido','data_envio']].head()
```

```python
# Realizando uma consulta para verificar  data Max e data Min 

data_pedido_max = df['data_pedido'].max()
data_pedido_min = df['data_pedido'].min()
data_envio_max = df['data_envio'].max()
data_envio_min = df['data_envio'].min()

print(f'Data Pedido : \n Maior data {data_pedido_max} \n Menor data: {data_pedido_min} \nData Envio: \n Maior data: {data_envio_max} \n Menor data: {data_envio_min}')
```

```python
# Criando o datagrame calendario utilizando o range do data max e data min 

calendario = pd.DataFrame({
    'data': pd.date_range(start=data_pedido_min, end=data_envio_max, freq='D')
})
```

```python
# Criando colunas para a calendario, coloquei algumas colunas que são utilizadas no dia-dia do BI

calendario['sk_data'] = calendario['data'].dt.strftime('%Y%m%d').astype(int)
calendario['ano'] = calendario['data'].dt.year
calendario['mes'] = calendario['data'].dt.month
calendario['dia'] = calendario['data'].dt.day
calendario['dia_semana'] = calendario['data'].dt.day_name(locale='pt_BR')
calendario['mes_nome'] = calendario['data'].dt.month_name(locale='pt_BR')
calendario['semana_ano'] = calendario['data'].dt.isocalendar().week.astype(int)
calendario['trimestre'] = calendario['data'].dt.quarter
calendario['semestre'] = ((calendario['data'].dt.month - 1) // 6) + 1
calendario['ano_mes'] = calendario['data'].dt.to_period('M').astype(str)   # ex: 2025-10
calendario['ano_trimestre'] = calendario['ano'].astype(str) + '-T' + calendario['trimestre'].astype(str)
calendario['ano_semestre'] = calendario['ano'].astype(str) + '-S' + calendario['semestre'].astype(str)
calendario['e_fim_de_semana'] = calendario['data'].dt.dayofweek >= 5
calendario['dia_do_ano'] = calendario['data'].dt.dayofyear
calendario = calendario.sort_values('data').reset_index(drop=True)
```

```python
# Remove a parte da hora de datetime, mantendo apenas a data
# Isso evita que o PostgreSQL insira '00:00:00' no timestamp

calendario['data'] = calendario['data'].dt.date
```

```python
# Novamente um top 5 só para conferir

calendario.head()
```

```python
# Ajustando as ordens da coluna
df_calendario = calendario[['sk_data','data','ano','mes','dia','dia_semana','mes_nome','semana_ano', 
                            'trimestre','semestre','ano_mes','ano_trimestre','ano_semestre', 
                            'e_fim_de_semana','dia_do_ano']]
 
```

```python
# Realizando uma nova conferencia visualizando um top 5

df_calendario.head()
```

```python
# Abre conexão e limpa completamente a tabela no banco, reiniciando o ID e removendo dependências
schema = "dw"
tabela = "dim_calendario"

with engine.begin() as conn:
    conn.execute(text(f'TRUNCATE TABLE {schema}.{tabela} RESTART IDENTITY CASCADE'))
```

```python
# Salvando os dados na tabela criado via banco

with engine.begin() as conn:
    df_calendario.to_sql(
        "dim_calendario",  # nome da tabela 
        conn,               # conexão com o banco de dados
        schema="dw",        # schema onde a tabela será criada
        if_exists="append", # substitui a tabela se já existir (garante dados atualizados)
        index=False,         # não cria coluna de índice do pandas
        method="multi",      # insere várias linhas por comando INSERT (melhor performance)
        chunksize=5000       # envia os dados em blocos de 5000 linhas (controle de memória)
    )
```

```python
# para verificarmos quantas linhas subiram

count = pd.read_sql("SELECT COUNT(*) as total_registros FROM dw.dim_calendario", engine).iloc[0,0]
print(f"Total de registros na tabela: {count}")
```
